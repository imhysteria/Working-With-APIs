# API 2 - AFL Teams

```python
# This request was initially blocked and required a head with the purpose to avoid request rejection
```

```python
headers = {'User-Agent': 'MyScript/1.0'}

response2 = requests.get('https://api.squiggle.com.au/?q=teams;format=json', headers=headers)
print("Status Code:", response2.status_code)
```

```python
results2 = response2.json()
results2
```

```python
pd.set_option("display.max_rows",0)
afl = pd.DataFrame(results2)
print(afl.shape)
display(afl.head(10))
```

```python
def flatten(afl, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(afl[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        afl = pd.concat([afl, flattened], axis=1)
        afl = afl.drop(column, axis=1)
    return afl
```

```python
afl['teams']
dict(afl['teams'])
pd.DataFrame(dict(afl['teams']))
afl_df = pd.DataFrame(dict(afl['teams'])).transpose()
display(afl_df)
```

```python
afl_df = afl_df[['id','abbrev','name','debut','logo']]
display(afl_df)
```
